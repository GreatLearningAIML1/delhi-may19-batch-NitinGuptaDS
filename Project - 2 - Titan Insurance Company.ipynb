# Solution: 
Probability of type 1 error = Level of significance of the test = 0.05 or 5%

When the null hypothesis is true and you reject it, you make a type I error. The probability of making a type I error is α, which is the level of significance you set for your hypothesis test. An α of 0.05 indicates that you are willing to accept a 5% chance that you are wrong when you reject the null hypothesis. To lower this risk, you must use a lower value for α. However, using a lower value for alpha means that you will be less likely to detect a true difference if one really exists.

# Question 4(b): What is the p- value of the hypothesis test if we test for a difference of $5000? (10 points)

As it is suggested that Titan will break even if the average output increases by 5K£. 
We will test this claim.
Our null hypothesis will be that an increase in output by 5K will not have a significant impact on the mean.

H0: Mu2-Mu1<=5000

H1: Mu2-Mu1 >5000

We will calculate t critical value for 5% significance level and 29 degrees of freedom. As we are testing for right tail (<=5000) we will use (1-alpha) = (1-0.005) for calculating area under the curve to the left of the "critical" value. 

Using t critical value we can calculate confidance interval. If 5000 is bigger than the critival value then its fall in the "region of rejection", hence null hypothesis can be rejected. if $5000 is lesser than the critical value, that means it falls in the "region of non rejection".

Subsequently, we will also calculate p value for the value of mean difference 5000 by:

1) dividing it with standard error for getting the t stat for mean diffenece of 5000

2) then calculating the p value for this t stat 

```python
#Calculate difference between pre and post treatment sample scores
diff = sp['NewScheme'] - sp['OldScheme ']

#Calculate degrees of freedom of sample
n=diff.count()
deg_frdm = n-1
print('Degrees of freedome is {}'.format(deg_frdm))

#Calculate standard error of diffence of sample scores
std_err = stdev(diff)/sqrt(n)
print('Standard Error of the test is {}'.format(std_err))

# Significance level
alpha = 0.05

# Calculate t_critical for alpha = 0.05 and df = n-1, actully we will have to use 1-alpha as this is a right til test.
t_critical = stats.t.ppf(1-alpha,n-1)
print('t ctitical value for significance level {} and {} degree of freedome is {}'.format(alpha, deg_frdm, t_critical))

#Calulate the differnec of means at t_critical
#As we have already taken diffenece of means, we can take muo = 0
mu0  = 0
xbar = mu0 + t_critical*std_err
print('Differnce of means for t ctitical is {}'.format(xbar))
```

As the critical value for "difference of means" is 4.3681
Since data originally is in 1000 hence : 4.3681 X 1000 = 4368.17615
Mean diiference of 5000 is to the right of this critical value, meaning it falls in the region of rejection.
Null hypothesis can be rejected, infering that an average raise of $5000 will significantly raise the output.


# Now we will also calculate the p value for mean difference of 5000, as required by the question

```python
#Calulate test statistic for differnece of means = 5000, 
# however to normalize in accordance to data 5000/1000 = 5
#meaning (xbar = 5000) = mu0 + t_stat*std_err, where we know that mu0 = 0
#so t_stat for (xbar = 5000) = xbar /std_err, lets calculate this
t_stat = 5/std_err
print('Test Statistics for mean differnec of 5000 is {}'.format(t_stat))

#Calculate the p value for test statistic, whole area under the right tail from t_stat for (xbar = 5000)
print('p value for mean differnece of 5000 is {} '.format(1-stats.t.cdf(t_stat,n-1)))
```

p value for the average gain for 5000 is 0.030771084315083463, it is less than 0.05 and in the region of rejection.

# Question 4(c): Power of the test (5 points)

The probability of correctly rejecting H0 when it is false is known as the power of the test. The larger it is, the better.

# Calculating Power of Test for the experiment with old scheme and new scheme

```python
# 1. Compute the difference in Means between 2 sample means 
diff_samp_means = NewSchMean - OldSchMean
print('Differnce of the sample means is {}'.format(diff_samp_means))

#2. Calculate Pooled Standard Deviation, as the 2 samples size are same we can use formula Sd_pool = sqrt((Var1 + var2)/2) 
Pooled_std_err = np.sqrt((np.var(sp['OldScheme '])+np.var(sp['NewScheme']))/2)
print('Pooled Standard Error is {}'.format(Pooled_std_err))

#3. Cohen's d is the measure of effect size, we will devide difference of means  by pooled std error for effect size
effect_size = diff_samp_means / Pooled_std_err
print('Effect Size is {}'.format(effect_size))

#4. number of Observations
nobs = sp['OldScheme '].count()
# Set the alpha value to 0.05 and alternative values 'two-sided' , 'larger' , 'smaller'
alpha = 0.05
alt = 'larger'

#5. Calculate power of test, as we are using one sided test, aternative is 'larger'
print('Power of the test is {}'.format(ttest_power(effect_size, nobs=nobs, alpha=alpha, alternative=alt)))
```
