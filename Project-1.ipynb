Write a function with name Goal_diff_count which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference. (5 points)

Goal_diff_count = GoalsFor - GoalsAgainst

```python
def Goal_diff_count( GoalsFor, GoalsAgainst ):   
    Goal_diff_count = GoalsFor - GoalsAgainst
    return Goal_diff_count;

result = Goal_diff_count(lg['GoalsFor'].astype(int), lg['GoalsAgainst'].astype(int))
print(result.min(),result.max())
```

```python
lg['Goal_diff_count'] = result
lg[lg.Goal_diff_count == lg.Goal_diff_count.max()]
```

```python
lg[lg.Goal_diff_count == lg.Goal_diff_count.min()]
```
